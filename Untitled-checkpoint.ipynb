```python
class employee:
     def __init__(self,a,b):
        self.id = a
        self.name = b
class salaried(employee):
    def __init__ (self,a,b,c):
        self.sal = c
        super().__init__ (a,b)
    def salary (self):
        return self.sal
class free(employee):
    def __init__ (self,a,b,c,d):
        self.hours = c
        self.rate = d
        super().__init__ (a,b)
    def salary (self):
        return self.hours*self.rate
class commission(salaried):
    def __init__ (self,a,b,c,com):
        self.com = com
        super().__init__(a,b,c)
    def salary (self):
        return self.sal+self.com
s=salaried(123,"abhi",100)
f=free(124,"golu",135,50)
c=commission(127,"madhab",7000,1950)

class salpro:
    def pro (self,t):
        for i in t:
            print(i.id,i.name,i.salary())
sa=salpro()
sa.pro([s,f,c])
        
```

```python
#1.WAP Implement a Student class with information such as rollno, name, class. The information must be entered by the user.

class Student_Details:
    def __init__(self):
        self.name = input("Enter Your Name: ")
        self.cname= input("Enter your college name: ")
        self.c = input("Enter your class: ")
        self.roll = int(input("Enter your Roll.No: "))

class Temp(Student_Details):
    def display(self):
        print("=======Student Details====== ")
        print("Name:", self.name)
        print("College Name: ", self.cname)
        print("Class: ", self.c)
        print("Roll Number: ", self.roll)

obj1 = Temp()
obj1.display()
```
